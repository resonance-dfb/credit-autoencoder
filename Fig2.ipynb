```python
import scienceplots
import torch
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt


plt.rcParams['font.serif'] = ['Times New Roman']
mpl.rcParams['figure.dpi'] = 800
mpl.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
plt.style.use(['science','ieee'])

# Define the sigmoid function and its derivative
def sigmoid(x, sigma):
    return x / (1 + torch.exp(-x / sigma))

def sigmoid_derivative(x, sigma):
    sigmoid_x = torch.sigmoid(x/sigma)
    return sigmoid_x + x * sigmoid_x * (1 - sigmoid_x)/sigma

# Define the ReLU derivative
def relu_derivative(x):
    return torch.where(x > 0, torch.tensor(1.0), torch.tensor(0.0))

# Generate x values
x = torch.linspace(-20, 20, 1000)

# Compute sigmoid values and derivatives for sigma=0.1 and sigma=10
sigma_1 = 1
sigma_2 = 5

sigmoid_values_1 = sigmoid(x, sigma_1)
sigmoid_values_2 = sigmoid(x, sigma_2)
derivative_values_1 = sigmoid_derivative(x, sigma_1)
derivative_values_2 = sigmoid_derivative(x, sigma_2)

# Plot for Fig2b
plt.figure(figsize=(5, 4))
# Replace horizontal line with ReLU derivative
plt.plot(x.numpy(), relu_derivative(x).numpy(), label=r'$\sigma = 0$', color='green', linestyle='-.', linewidth=1)
plt.plot(x.numpy(), derivative_values_1.numpy(), label=r'$\sigma = 1$', color='blue', linestyle='-', linewidth=1)
plt.plot(x.numpy(), derivative_values_2.numpy(), label=r'$\sigma = 5$', color='red', linestyle='--', linewidth=1)

plt.xlabel('x', fontsize=15)
plt.ylabel(r'$d\Phi_\xi(x;\sigma)/dx$', fontsize=15)
plt.legend(fontsize=15, loc='upper left')
plt.tick_params(axis='both', which='major', labelsize=15)

# Add label (b) in the upper-left corner
plt.text(-26, 1.15, '(d)', fontsize=17, verticalalignment='top', horizontalalignment='right')
# Save the figure (second plot)
plt.savefig('D:/Credit Card Fraud Detection/Fig2d.eps', dpi=800)
plt.show()
```
